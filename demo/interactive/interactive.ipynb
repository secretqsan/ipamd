```python
from galadrop import App
from galadrop.public.md_common import Environment
```

Step 2: create a simulation app and use Calvados3 force field

```python

app = App(name = 'test').use('Calvados3')
```

step 3: create a simulation box in specified environment

```python
box = app.builder.new_simulation_box(20, 20, 20)
environment = Environment()
environment.set_ph(7)
environment.set_temperature(295.15)
environment.set_ionic_strength(0.15)
box.in_solvent(environment)
```

Step 4 : create an ELP and locate it in the simulation box

```python
app.builder.af2('protein1', 'VPGRGVPGRGVPGRGVPGRGVPGRGVPGRGVPGRGVPGRG')
mol1 = app.builder.protein_from_pdb('protein1.pdb', rigid_from_plddt=True)
box.place_molecule_randomly(mol1, 1)
```

Step 5: run simulation!

```python
simulation = app.simulation.new_simulation(box, 'simulation', total_time='1ns', snap_shot=20)
simulation.run()
```

Step 6: view simulation snapshots

```python
box.view()
```

```python
import py3Dmol

p = py3Dmol.view(query='mmtf:1ycr')
p.setStyle({'cartoon': {'color':'red'}})
```

Step 7: analyse propertities of the system

```python
density = box.compute(app.analysis.density, title='density', lx=10, ly=10, lz=10)
density.print()
```
